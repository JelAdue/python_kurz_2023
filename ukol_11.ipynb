Stáhni si soubor platy_2021_02.csv s informacemi o platech v softwarové firmě (stejný soubor, jako v předchozím úkolu).

Načti si tato data do tabulky a vytvoř histogram. Nastav vhodně hranice skupin histogramu (parametr bins), aby byl graf přehledný a snadno interpretovatelný.

```python
import pandas
```

```python
#načtení dat a podívání se na ně
platy_unor = pandas.read_csv("platy_2021_02.csv")
#platy_unor
```

```python
#zjistila jsem si kolik je max a min, abych vedela, jak stanovit prvni a posledni hranici
print (platy_unor.max())
print (platy_unor.min())
```

```python
# po 5000
# ještě málo přehledné, v dalším to zkusím ještě rozpůlit
platy_unor["plat"].plot(kind="hist",bins=[30000, 35000, 40000, 45000, 50000, 55000, 60000],grid=True)
```

```python
# hranice po 2500
# je to lepší, ale zkusím to ještě jednou rozpůlit
platy_unor["plat"].plot(kind="hist",bins=[30000, 32500, 35000, 37500, 40000, 42500, 45000, 47500, 50000, 52500, 55000, 57500, 60000],grid=True)
```

```python
# hranice po 1250
platy_unor["plat"].plot(kind="hist",bins=[30000, 31250, 32500, 33750, 35000, 36250, 37500, 38750, 40000, 41250, 42500, 43750, 45000, 46250, 47500, 48750, 50000, 51250, 52500, 53750, 55000, 56250, 57500, 58750, 60000],grid=True)
```

```python
# no a potom přišel nápad, že po 1000 to asi bude nejlepší :D a že se mi to nechce vypisovat,takže to pujde určitě nějak naprogramovat... hmm a to už jsme někde dělali..

platy_unor["plat"].plot(kind="hist",bins=range(30000,60000,1000),grid=True)
```

```python
# nedalo mi to ale ještě zkusit 500, ale to už mi příjde nepřehledné
platy_unor["plat"].plot(kind="hist",bins=range(30000,60000,500),grid=True)
```
